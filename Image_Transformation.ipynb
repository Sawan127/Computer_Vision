Importing libraries

```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage.io import imread,imshow
```

Read image


```python
image = plt.imread('/content/drive/MyDrive/Colab Notebooks/IPCV/sunflower.jpg')
plt.imshow(image)
plt.show()
```

1. Splitting image into BGR then merging in RGB format

```python
b,g,r = cv2.split(image) # split image
```

```python
img = cv2.merge((r,g,b)) # merge image
```

```python
imshow(img) # RGB image
plt.title('RGB image')
```

RGB to Negative : cv2.bitwise_not(image)

```python
plt.subplot(1,2,1),plt.imshow(img),plt.title('Original')

img_neg = cv2.bitwise_not(img) # Negative image

plt.subplot(1,2,2),plt.imshow(img_neg),plt.title('Negative')
plt.show()
```

RGB TO GRAYSCALE

```python
gray_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
imshow(gray_image)
```

RGB TO HSV

```python
hsv_img = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
imshow(hsv_img)
```
