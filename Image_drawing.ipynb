```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread,imshow
%matplotlib inline
```

```python
#  Create a blank image
# We create the canvas to draw: 400 x 400 pixels, 3 channels, uint8 (8-bit unsigned integers)
# We set background to black using np.zeros()

img = np.zeros((400,400,3),dtype= np.uint8) # black image
imshow(img)
```

```python
# Change Background of image
img[:] = [125,125,125] # Gray
imshow(img)
```

Drawing Rectangle

The signature for the cv2.rectangle() function is as follows:

img = rectangle(img, pt1, pt2, color, thickness=1, lineType=8, shift=0)


```python
rectangle = cv2.rectangle(img,(30,25),(100,50),[255,255,0],2) #thickness is 2
rec = cv2.rectangle(img, (65,65),(200,250),[255,0,255],-1) # filled rectangle with color
plt.imshow(rec)
plt.show()
# plt.imshow(rec)
# plt.show()
```

```python
# Square
square = cv2.rectangle(img,(150,150),(300,300),[0,0,0],4)
imshow(square)
```

Drawing Circles

The signature for the cv2.circle() function is as follows:

img = circle(img, center, radius, color, thickness=1, lineType=8, shift=0)


```python
cir = cv2.circle(img, (325,50),50,[0,255,255],-1)
imshow(img)
```

Drawing Ellipse

The signature for this function is as follows:

cv2.ellipse(img, center, axes, angle, startAngle, endAngle, color,
thickness=1, lineType=8, shift=0)


```python
ellip = cv2.ellipse(img,(50,350),(60, 40), 0, 0, 360,[0,0,255],-1)
plt.imshow(ellip)
```

```python
image=np.full((500,500,3),[255,255,255],dtype=np.uint8)
imshow(image)
plt.show()
```

Writing Text

The cv2.putText() function has the following signature:

img = cv.putText( img, text, org, fontFace, fontScale, color, thickness=1,
lineType= 8, bottomLeftOrigin=False)

```python
cv2.putText(image, 'OpenCV4 with Python', (10, 30),
cv2.FONT_HERSHEY_SIMPLEX, 0.9, [255,0,0], 2, cv2.LINE_4)
cv2.putText(image, 'OpenCV4 with Python', (10, 70),
cv2.FONT_HERSHEY_SIMPLEX, 0.9, [255,0,0], 2, cv2.LINE_8)
cv2.putText(image, 'OpenCV4 with Python', (10, 110),
cv2.FONT_HERSHEY_SIMPLEX, 0.9, [255,0,0], 2, cv2.LINE_AA)
plt.imshow(image)
```
