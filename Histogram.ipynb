```python
def show_grayscale_equalize(image):
  grayscale_image = cv2.cvtCOLOR(image,cv2.COLOR_BGR2GRAY)
  eq_grayscale_image = cv2.equalizeHist(grayscale_image)
  plt.imshow(eq_grayscale_image, cmap='gray')
  plt.show()
```

```python
#OpenCV provides the function cv2.equalizeHist to equalize the histogram of an image.
#The signature is the following:
#cv2.equalizeHist(image)

hist1_image=cv2.equalizeHist(gray_img)
res = np.hstack((gray_img,hist1_image)) #stacking images side-by-side
#cv2.imshow("original",gray_image)
#cv2.imshow("histoEqImage",hist1_image)
# cv2_imshow("Equalization-diff",res)
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
ClaheEqHist = clahe.apply(gray_img)
hist1_image=cv2.equalizeHist(gray_img)
res = np.hstack((gray_img,hist1_image,ClaheEqHist)) #stacking images side-by-side
cv2_imshow(res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
