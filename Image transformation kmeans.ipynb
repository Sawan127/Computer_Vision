```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# load the image
image = cv2.imread('/content/eeswar_photography1.png')
plt.imshow(image)
plt.show()
```

```python
# converting to RGB from BGR
img = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
# Reshaping the image into a 2D array of pixels and 3 color values (RGB)
pixel_value = image.reshape((1,-3))

# convert to float type
pixel_vals = np.float32(pixel_value)

# define the criteria to stop algorithm
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,100,0.85)

# define cluster as k =3
k =3
retval,labels,centers = cv2.kmeans(pixel_vals, k,None, criteria,10,cv2.KMEANS_RANDOM_CENTERS)


# convert data type to 8-bit values
centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

# reshape data into the original image dimensions
segmented_image = segmented_data.reshape((image.shape))

plt.imshow(segmented_image)
plt.show()
```
